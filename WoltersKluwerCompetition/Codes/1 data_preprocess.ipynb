```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
sample = pd.read_csv('sample_submission.csv')
```

```python
sample
```

```python
train.sort_values('id',inplace=True)
```

```python
train
```

```python
data = train.groupby(['id'])['cat_name'].apply(list).to_frame()

data['cat_name']=data['cat_name'].apply(lambda x:str(x).replace('[','').replace(']','').replace('\'',''))
data.reset_index(drop=True, inplace=True)
```

```python
data
```

```python
train_new = train.drop('cat_name', axis = 1)
train_new.drop_duplicates(inplace = True)
train_new.reset_index(drop=True,inplace=True)
```

```python
train_new
```

```python
final_train = pd.concat([train_new, data], axis = 1)
```

```python
final_train 
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
mlb_result = mlb.fit_transform([str(final_train.loc[i,'cat_name']).split(', ') for i in range(len(final_train))])
train_final = pd.concat([final_train[['id', 'name', 'document_text']],pd.DataFrame(mlb_result,columns=list(mlb.classes_))],axis=1)
```

```python
train_final
```

```python
train_final.columns.tolist()
```

```python
target_table = []
for index, row in train_final.iterrows():
    target = []
    for i in range(3,53):
        target.append(row[i])
    target_table.append([row['id'], target])
```

```python
target_table
```

```python
target_df = pd.DataFrame(target_table, columns = ['id', 'target'])
```

```python
train_with_target = pd.concat([final_train[['id', 'name', 'document_text']],target_df[['target']]],axis=1)
```

```python
train_with_target
```

```python
train_with_target.to_csv('train_with_target.csv')
```
